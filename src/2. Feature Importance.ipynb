```python
# load datasets
ph_tr = pd.read_csv("./Phising_Training_Dataset.csv")
ph_te = pd.read_csv("./Phising_Testing_Dataset.csv")

print(ph_tr.shape)
print(ph_te.shape)

ph_tr.head()
ph_te.head()
```

```python
# split into X and Y

ph_tr_ind = ph_tr.set_index('key')
X_tr = ph_tr_ind.drop('Result', axis = 1, inplace = False)
y_tr = ph_tr_ind.loc[:,['Result']]

print(X_tr.shape)
print(y_tr.shape)

X_tr.head()
```

```python
# train a decision tree on features to find important features

from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()

dt.fit(X_tr, y_tr)
```

```python
feat_imp_df = pd.DataFrame()

feat_imp_df['features'] = X_tr.columns
feat_imp_df['importance'] = dt.feature_importances_

feat_imp_df.sort_values('importance', ascending = False, inplace = True)
feat_imp_df.head(35)
```
