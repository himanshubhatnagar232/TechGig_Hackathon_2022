```python
# load datasets
ph_tr = pd.read_csv("./Phising_Training_Dataset.csv")
ph_te = pd.read_csv("./Phising_Testing_Dataset.csv")

print(ph_tr.shape)
print(ph_te.shape)

ph_tr.head()
ph_te.head()
```

```python
# split into X and Y

ph_tr_ind = ph_tr.set_index('key')
X_train = ph_tr_ind.drop('Result', axis = 1, inplace = False)
y_train = ph_tr_ind.loc[:,['Result']]

print(X_train.shape)
print(y_train.shape)
print(y_train[y_train['Result'] == 1].shape)
print(y_train[y_train['Result'] == -1].shape)
print('=' * 20)

ph_te_ind = ph_te.set_index('key')
X_test = ph_te_ind

print(X_test.shape)

X_train.head()
```

```python
# split dataset into training, validation and test

from sklearn.model_selection import train_test_split

X_tr, X_cv, y_tr, y_cv = train_test_split(X_train, y_train, test_size = 0.25, random_state = 123, stratify = y_train)

print(X_tr.shape)
print(X_cv.shape)
print(y_tr.shape)
print(y_cv.shape)

X_tr.head()
```

```python
from lightgbm import LGBMClassifier

lgb = LGBMClassifier()
lgb.fit(X_tr, y_tr['Result'])
```

```python
from sklearn.metrics import roc_curve, f1_score, confusion_matrix

# predict probability for validation data (for class 1)
y_cv_proba = lgb.predict_proba(X_cv)[:,1:]

# find roc auc score and thresholds
y_cv_norm = [0 if y == -1 else y for y in y_cv['Result']]
fpr, tpr, thresholds = roc_curve(y_cv_norm, y_cv_proba)

# calculate g-means of each threshold
gmeans = np.sqrt(tpr * (1 - fpr))

# find best threshold using gmeans
ix = np.argmax(gmeans)
print('Best Threshold=%f, G-Mean=%.3f' % (thresholds[ix], gmeans[ix]))

# find best threshold using f1_score
f1_sc_list = []
for th in thresholds:
  y_cv_pred = [1 if p > th else 0 for p in y_cv_proba]
  f1_sc = f1_score(y_cv_norm, y_cv_pred)
  f1_sc_list.append(f1_sc)

max_f1_ix = np.argmax(f1_sc_list)
print('Best Threshold=%f, F1-Score=%.3f' % (thresholds[max_f1_ix], f1_sc_list[max_f1_ix]))

# find best threshold using J statistic
J = fpr - tpr
ix_j = np.argmax(J)
print('Best Threshold=%f, J=%.3f' % (thresholds[ix_j], J[ix_j]))

# find F1 score at best GM threshold
best_thres = thresholds[ix]
y_cv_pred_gm = [1 if p > best_thres else 0 for p in y_cv_proba]
f1_sc_gm = f1_score(y_cv_norm, y_cv_pred_gm)
print('F1-Score (GM Threshold)=%.3f' % (f1_sc_gm))

# find F1 score at best J threshold
best_thres = thresholds[ix_j]
y_cv_pred_j = [1 if p > best_thres else 0 for p in y_cv_proba]
f1_sc_j = f1_score(y_cv_norm, y_cv_pred_j)
print('F1-Score (J Threshold)=%.3f' % (f1_sc_j))
```

```python
# plot confusion matrix

from sklearn.metrics import confusion_matrix

# for threshold using gm
print(confusion_matrix(y_cv_norm, y_cv_pred_gm))

# for threshold using f1 score
best_thres = thresholds[max_f1_ix]
y_cv_pred_f1 = [1 if p > best_thres else 0 for p in y_cv_proba]
print(confusion_matrix(y_cv_norm, y_cv_pred_f1))

# for threshold using j matrix
print(confusion_matrix(y_cv_norm, y_cv_pred_j))
```

```python
# generate CSV submission file

# get probabilities of class 1
y_test_proba = lgb.predict_proba(X_test)[:,1:]

# get prediction using ideal threshold
thresh_used = thresholds[max_f1_ix]
y_test_pred = [1 if p > best_thres else 0 for p in y_test_proba]

print(len(y_test_proba))

df_final_test = X_test.reset_index()
df_final_test['Result'] = y_test_pred

df_submission = df_final_test.loc[:,['key','Result']]

# save in CSV

import datetime
cur_date = datetime.datetime.now()
date_str = cur_date.strftime("%Y-%m-%d-%H%M%S")

filename = './Submissions/' + date_str + '_LGBM_predictions_akatsuki.csv'
print(filename)
df_submission.to_csv(filename, index = False)

df_submission.head()
```
